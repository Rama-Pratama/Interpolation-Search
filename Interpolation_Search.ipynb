```python
import csv
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import pandas as pd
import time
from PIL import ImageTk, Image
from tkinter import Label

def read_csv(file_path):
    data = []
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return sorted(data, key=lambda x: x.get('Place_Name', '').lower())

def interpolate_search(destination, data):
    low = 0
    high = len(data) - 1

    while low <= high and destination >= data[low]['Place_Name'] and destination <= data[high]['Place_Name']:
        pos = low + int(((high - low) / (ord(data[high]['Place_Name'][0]) - ord(data[low]['Place_Name'][0]))) * (ord(destination[0]) - ord(data[low]['Place_Name'][0])))

        if data[pos]['Place_Name'] == destination:
            return data[pos]
        elif data[pos]['Place_Name'] < destination:
            low = pos + 1
        else:
            high = pos - 1
  
    return None

def search_button_clicked():
    start_time = time.perf_counter()  # Start the timer
    
    destination = entry.get()
    result_text = ""
    
    matching_rows = df[df['Place_Name'] == destination]  # Assuming you are searching in the 'Place' column
    if not matching_rows.empty:
        for index, row in matching_rows.iterrows():
            tree.insert('', 'end', values=(row['Place_Name'], row['Description'], row['City'], row['Price'], row['Rating'], row['Category']))
    else:
        tree.insert('', 'end', values=("Not Found", "", "", "", "", "", ""))
    end_time = time.perf_counter()  # Stop the timer
    elapsed_time = end_time - start_time
    
    # Display the elapsed time in the GUI
    time_label.config(text=f"Interpolation Search Elapsed Time: {elapsed_time:.8f} detik")

df = pd.read_csv('Wisata.csv')

root = tk.Tk()
root.title("Interpolation Search Destinasi Wisata")

bg = ImageTk.PhotoImage(Image.open("Bg (1).jpg"))
my_label = Label(image=bg)
my_label.place(relwidth=1, relheight=1)

label = tk.Label(root, text="Nama Destinasi Wisata:")
label.pack(pady=10)

entry = tk.Entry(root)
entry.pack(pady=10)

search_button = tk.Button(root, text="Search",command=search_button_clicked)
search_button.pack(pady=10)

time_label = tk.Label(root, anchor='center', text="Interpolation Search Elapsed Time:")
time_label.pack(pady=10)

columns = ['Place_Name', 'Description', 'City', 'Price', 'Rating', 'Category']
tree = ttk.Treeview(root, columns=columns, show='headings')
tree.pack(pady=10)

for col in columns:
    tree.heading(col, text=col)
    tree.column(col, width=100)

root.geometry("800x600")
root.mainloop()
```
